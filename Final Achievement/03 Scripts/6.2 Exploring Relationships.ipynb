# Categorical Plots

Creating a categorical variable that splits the "Price per Unit" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df2['GDPCAP'], bins = 20, kde = True)
```

```python
df2.loc[df2['GDPCAP'] < 20, 'Price category'] = 'Low price'
```

```python
df2.loc[(df2['GDPCAP'] >= 20) & (df2['GDPCAP'] < 50), 'Price category'] = 'Middle price'
```

```python
df2.loc[df2['GDPCAP'] >= 50, 'Price category'] = 'High price'
```

```python
df2['Price category'].value_counts(dropna = False)
```

```python
df2.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="GDPCAP", y="HDI", hue="Price category", data=df2)
```
